```python
import sys
sys.path.append("../")
import pymongo
import time
from lib.db import MongoClient
import csv
import datetime

c = MongoClient("mongodb://localhost:27017")
```

```python
updates = []
with open("../Avg repair times.csv", newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    counter = 0
    for row in spamreader:
        counter += 1
        if counter == 1:
            continue
        repair_time = row[0]
        emu = row[1]
        try:
            x = time.strptime(repair_time, '%H:%M:%S')
        except:
            continue
        second = int(datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds())
        if second != 0:
            print(emu, second)
            
            updates.append(pymongo.UpdateOne({"lifts": {"$in": [emu]}}, {"$addToSet": {"repair_times": second}}))
```

```python
print(f">> adding {len(updates)} repair times")
c.stations.bulk_write(updates)
```
